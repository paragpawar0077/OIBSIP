```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
unemp_df = pd.read_csv("Unemployment in India.csv")
```

```python
unemp_df.shape
```

```python
unemp_df.columns
```

```python
unemp_df.columns = unemp_df.columns.str.strip()
```

```python
unemp_df.head(10)
```

```python
unemp_df.info()
```

```python
unemp_df.isna()
```

```python
unemp_df.isna().sum()
```

```python
unemp_df = unemp_df.dropna(how='all')
```

```python
unemp_df.shape
```

```python
unemp_df.isna().sum()
```

```python
unemp_df.info()
```

```python
unemp_df["Date"] = pd.to_datetime(unemp_df["Date"])
```

```python
unemp_df.info()
```

```python
unemp_df["Year"] = unemp_df["Date"].dt.year
unemp_df["Month"] = unemp_df["Date"].dt.month
```

```python
unemp_df.head()
```

```python
monthly_avg = unemp_df.groupby("Date")["Estimated Unemployment Rate (%)"].mean()

plt.figure(figsize=(12,6))
plt.plot(monthly_avg)
plt.title("Overall Unemployment Trend in India")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
covid_2020 = unemp_df[unemp_df["Year"] == 2020]

covid_avg = covid_2020.groupby("Date")["Estimated Unemployment Rate (%)"].mean()

plt.figure(figsize=(12,6))
plt.plot(covid_avg)
plt.title("Unemployment Trend During 2020 (Covid Period)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
area_avg = unemp_df.groupby("Area")["Estimated Unemployment Rate (%)"].mean()

area_avg.plot(kind="bar")
plt.title("Average Unemployment Rate: Rural vs Urban")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
state_avg = unemp_df.groupby("Region")["Estimated Unemployment Rate (%)"].mean()

top_states = state_avg.sort_values(ascending=False).head(10)

top_states.plot(kind="bar")
plt.title("Top 10 States with Highest Unemployment")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

# Conclusion

In this project, we analyzed unemployment trends in India using historical data. The analysis showed noticeable fluctuations in unemployment rates over time, with a significant rise during the COVID-19 period in 2020.

The comparison between different regions and areas (Rural and Urban) revealed that unemployment levels vary across states and locations. Some regions were more affected than others, highlighting regional economic differences.

Overall, the study demonstrates how major events like the pandemic can strongly impact employment levels and emphasizes the importance of monitoring unemployment trends for better economic planning.
